```python
AVL = []
BST = []
name = "delete_ordered"
with open(name+".txt") as file:
    lines = file.read().split("\n")
    for line in lines:
        if(len(line)==0): continue
        tokens = line.split(" ")
        lis = []
        if(tokens[0]=="BST"):
            lis = BST
        else:
            lis = AVL
        lis.append((int(tokens[1]),int(tokens[2])))
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create data
bst_x = [p[0] for p in BST]
bst_y = [p[1] for p in BST]
avl_x = [p[0] for p in AVL]
avl_y = [p[1] for p in AVL]
  
# plot lines
plt.plot(bst_x, bst_y, label = "BST")
plt.plot(avl_x, avl_y, label = "AVL")
plt.ylabel("Runtime in ms")
plt.xlabel("Number of items")
plt.legend()
# plt.show()
plt.savefig(name+".jpg")
```
